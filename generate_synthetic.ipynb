```python
import os
import random
from datetime import datetime, timedelta
import pandas as pd
import numpy as np

DATA_DIR = "data"
RAW_PATH = os.path.join(DATA_DIR, "raw.csv")

np.random.seed(42)
random.seed(42)

os.makedirs(DATA_DIR, exist_ok=True)

# Parámetros
N = 500  # filas

first_names = ["Ana", "Luis", "María", "Carlos", "Lucía", "Pedro", "Sofía", "Jorge", "Elena", "Diego"]
last_names = ["Pérez", "García", "Rodríguez", "Sánchez", "López", "Torres", "Ramírez", "Flores", "Díaz", "Castro"]
domains = ["example.com", "mail.com", "work.org", "contoso.pe"]

start_date = datetime(2023, 1, 1)

rows = []
for i in range(N):
    uid = i + 1
    fn = random.choice(first_names)
    ln = random.choice(last_names)
    name = f"{fn} {ln}"

    # Email (a propósito con errores en ~10%)
    if random.random() < 0.1:
        email = f"{fn.lower()}{ln.lower()}@{random.choice(domains)}".replace("@", "")  # inválido sin @
    else:
        email = f"{fn.lower()}.{ln.lower()}@{random.choice(domains)}"

    # Monto puede venir como string con símbolos, o NaN
    amount = round(abs(np.random.normal(50, 20)), 2)
    if random.random() < 0.1:
        amount_str = f"S/. {amount}"  # con símbolo
    elif random.random() < 0.05:
        amount_str = ""  # vacío
    else:
        amount_str = str(amount)

    # Fecha con posibles formatos/errores
    d = start_date + timedelta(days=random.randint(0, 540))
    if random.random() < 0.08:
        date_str = d.strftime("%d-%m-%Y")  # formato distinto
    elif random.random() < 0.03:
        date_str = "32/13/2023"  # inválida
    else:
        date_str = d.strftime("%Y-%m-%d")

    # Duplicados intencionales (~5%)
    if random.random() < 0.05 and rows:
        rows.append(random.choice(rows))
        continue

    rows.append({
        "user_id": uid,
        "name": name,
        "email": email,
        "amount": amount_str,
        "created_at": date_str,
        "country": random.choice(["PE", "PE ", "pe", "MX", "CO", "AR", None])
    })

# Agregamos algunos tipos incorrectos a propósito
for _ in range(5):
    rows.append({
        "user_id": "NA",
        "name": 12345,
        "email": "sin-correo",
        "amount": "S/. -",
        "created_at": "ayer",
        "country": "??"
    })

pd.DataFrame(rows).to_csv(RAW_PATH, index=False)
print(f"Archivo crudo generado en {RAW_PATH}")
```
